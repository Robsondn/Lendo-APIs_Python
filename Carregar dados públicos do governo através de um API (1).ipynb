####  2. Melhorando a interação com o API
Observe dois URLs de consultas diferentes. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1. Meu código aqui

import requests
import pandas as pd
from IPython.display import display

def consulta_pvl(uf, tipo_interessado):
    """
    Realiza uma consulta na API de PVL com base na UF e tipo de interessado.

    Args:
        uf (str): A Unidade Federativa para filtrar os dados (ex: 'RJ', 'MG', 'BA').
        tipo_interessado (str): O tipo de interessado (ex: 'Estado' ou 'Município').

    Returns:
        pandas.DataFrame: DataFrame com os dados da consulta, ou um DataFrame vazio se ocorrer um erro.
    """
    url = "https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl"
    try:
        response = requests.get(url)
        response.raise_for_status()
        data = response.json()
        df = pd.DataFrame(data['items'])
        df_filtered = df[(df['uf'] == uf) & (df['tipo_interessado'] == tipo_interessado)]
        return df_filtered
    except requests.exceptions.RequestException as e:
        print(f"Erro na requisição: {e}")
        return pd.DataFrame()
        
````

````python
# 2. Meu código aqui

df_mg_estado = consulta_pvl('MG', 'Estado')
if not df_mg_estado.empty:
  arquivados_mg = df_mg_estado[df_mg_estado['status'] == 'Arquivado por decurso de prazo']
  num_arquivados = len(arquivados_mg)
  print(f"Número de solicitações arquivadas por decurso de prazo em Minas Gerais: {num_arquivados}")
else:
  print("Não foi possível recuperar dados para Minas Gerais.")
````

````python
# 3. Meu código aqui

df_ba_municipio = consulta_pvl('BA', 'Município')
if not df_ba_municipio.empty:
    df_ba_deferidos = df_ba_municipio[df_ba_municipio['status'] == 'Deferido']
    if not df_ba_deferidos.empty:
      municipio_mais_deferido = df_ba_deferidos['interessado'].value_counts().idxmax()
      print(f"O município da Bahia com mais solicitações deferidas é: {municipio_mais_deferido}")
    else:
      print("Não há solicitações deferidas para municípios da Bahia.")
else:
  print("Não foi possível recuperar dados para municípios da Bahia.")
````

````python
# 4. Meu código aqui

df_ba_estado = consulta_pvl('BA', 'Estado')
if not df_ba_estado.empty:
    df_ba_estado.to_csv('solicitacoes_bahia_estado.csv', index=False)
    print("Dados das solicitações do estado da Bahia salvos em 'solicitacoes_bahia_estado.csv'")
else:
    print("Não foi possível recuperar dados para o estado da Bahia.")
````
